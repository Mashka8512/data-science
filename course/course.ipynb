```python
# евклидово расстояние
def predict(X_train, y_train, x, X_test, y_test, k):
    
    index=np.zeros(k)
    pred=np.zeros(len(y_test))
    for i in range(len(y_test)):
        dist=np.zeros((len(y_train),2))
        #print("begin: ",i)
        for j in range(len(y_train)):
            n=min(X_train[j][1]-X_train[j][0], X_test[i][1]-X_test[i][0])
            #print( "n: ",n)
            
            for t in range(n):
                dist[j][0]+=(x["Open price"][X_train[j][1]-t]-x["Open price"][X_test[i][1]-t])**2
                dist[j][0]+=(x["Min"][X_train[j][1]-t]-x["Min"][X_test[i][1]-t])**2
                dist[j][0]+=(x["Max"][X_train[j][1]-t]-x["Max"][X_test[i][1]-t])**2
                dist[j][0]+=(x["Close price"][X_train[j][1]-t]-x["Close price"][X_test[i][1]-t])**2
            dist[j][0]=sqrt(dist[j][0])
            dist[j][1]=j  
            #print("dist: ", dist[j][0])
            #if dist[j][0]<50:
                #print ("yes: ",j, " ", y_train[j], " ", dist[j][0])
        np.min(dist, axis=0) 
        sum1=0
        sum2=0
        for j in range(k):
            index[j]=dist[np.argmin(dist, axis=0)[0], 1]
            
            #print ("ind: ", index[j], " ", dist[np.argmin(dist, axis=0)[0], ])
            #print(dist)
            
            if y_train[int(index[j])]==1:
                #print("1: ")
                sum1=sum1+1
            else:
                sum2=sum2+1
            dist=np.delete(dist, dist[np.argmin(dist, axis=0)[0],], axis=0)
            #print(dist)
        if sum1>sum2:
            pred[i]=1
        else:
            pred[i]=-1
    return pred

```
